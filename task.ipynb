```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


df = pd.read_csv("/content/Wholesale customers data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Step 3: Select features for clustering
features = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
X = df[features]

# Step 4: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 5: Choose the number of clusters (k)
k = 3

# Step 6: Initialize the k-means algorithm
kmeans = KMeans(n_clusters=k, random_state=42)

# Step 7: Fit the model to the data
kmeans.fit(X_scaled)

# Step 8: Get the cluster assignments
cluster_labels = kmeans.labels_

# Step 9: Visualize the clusters (optional)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_scaled[:, 0], X_scaled[:, 1], X_scaled[:, 2], c=cluster_labels, cmap='viridis', s=50)

ax.set_xlabel('Fresh')
ax.set_ylabel('Milk')
ax.set_zlabel('Grocery')

plt.title('K-Means Clustering with 3 clusters')
plt.show()
```
